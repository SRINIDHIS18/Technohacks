TASK1:PERFORM DATA CLEAN


Clean a dataset by removing missing values
and outliers

```python
import numpy as np
import pandas as p
import matplotlib.pyplot as pl
import seaborn as sb
```

```python
df=p.read_csv('/content/titanic.csv')
df.isna().sum()
sb.boxplot(df)
```

```python
sb.boxplot(df['Age'])
```

```python
df['Age']=df['Age'].clip(lower=df['Age'].quantile(0.07),upper=df['Age'].quantile(0.97))
sb.boxplot(df['Age'])
```

```python
sb.boxplot(df['Fare'])
```

```python
df['Fare']=df['Fare'].clip(lower=df['Fare'].quantile(0.10),upper=df['Fare'].quantile(0.95))
sb.boxplot(df['Fare'])
```
